```python
import numpy as np
import matplotlib.pyplot as plt

def numerical_diff(f,x):
    h = 1e-4
    return (f(x+h)-f(x-h))/(2*h)
def func_1(x):
    return 0.01*x**2+0.1*x
```

```python
x = np.arange(1.0,20.0,0.1)
y = func_1(x)
s = []
for n in range(len(x)):
    s.append(numerical_diff(func_1,n))

plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,y)
plt.plot(x,s,'r--')
plt.show()

```
