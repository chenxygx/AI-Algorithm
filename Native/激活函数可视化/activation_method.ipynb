```python
import numpy as np
from matplotlib.pylab import plt

plt.rcParams['font.sans-serif'] = ['KaiTi']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def step_function(x):
    return np.array(x>0,dtype=np.int)

def relu(x):
    return np.maximum(0,x)

def sigmoid(x):
    return 1.0/(1+np.exp(-x))

def identity(x):
    return x
```

```python
x = np.arange(-5,5,0.01)
y = step_function(x)
plt.plot(x,y,'y-',label="阶段函数")

y_s = sigmoid(x)
plt.plot(x,y_s,'r--',label="sigmoid")

y_r = relu(x)
plt.plot(x,y_r,'b-.',label="ReLU")

y_i = identity(x)
plt.plot(x,y_i,'g--',label="恒等函数")

plt.legend(title="激活函数")
plt.ylim(-0.4,1.5)
plt.show()
```
