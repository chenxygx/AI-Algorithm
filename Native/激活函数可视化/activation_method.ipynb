```python
import numpy as np
from matplotlib.pylab import plt

def step_function(x):
    return np.array(x>0,dtype=np.int)

def sigmoid(x):
    return 1.0/(1+np.exp(-x))

def relu(x):
    return np.maximum(0,x)

x = np.arange(-5,5,0.01)
y = step_function(x)
plt.ylim(-0.1,1.5)
plt.plot(x,y,'y-',label="step")

y_s = sigmoid(x)
plt.plot(x,y_s,'r--',label="sigmoid")

y_r = relu(x)
plt.plot(x,y_r,'b-.',label="ReLU")

plt.grid(axis='x', color='0.95')
plt.legend(title="type")
plt.title("ActivationMethod")
plt.show()
```
